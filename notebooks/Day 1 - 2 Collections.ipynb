## Dictionaries

Another - yet intuitive - collection is a `dictionary`.

The important difference here to a list and a tuple, is the importance of a __key - value__ pair

```python
kv = (('name', 'anki'),('age', 99), ('affiliation','FAU'))
```

```python
d = dict(kv)
```

```python
d
```

```python
d['age'] = 5
```

```python
d
```

```python
d['shoesize'] = 43
```

```python
d
```

```python
A = ["name","age","affiliation"]
B = ["anki",5,"FAU"]

d = dict(zip(A, B))

d
```

Okay, when writing `dict(my_dictionary)`, the first element was `age`, can I access it when I have the key `0`?

```python
d.keys()
```

```python
d.values()
```

```python
d.items()
```

```python
d.pop('name')
```

```python
d
```
