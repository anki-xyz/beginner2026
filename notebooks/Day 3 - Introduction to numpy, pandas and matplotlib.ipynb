### Example 2: Difference between numpy arrays and Python lists

```python
# Python list operations
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)  # Concatenates the lists

# numpy array operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # Element-wise addition
```

## 2. pandas Introduction
**pandas** is a popular data manipulation library in Python. It provides data structures like Series and DataFrame for handling and analyzing structured data.

### Example 1: Creating a DataFrame

```python
import pandas as pd

# Create a simple dataframe
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "My column": ["a", "b", "c"]
})

df
```
