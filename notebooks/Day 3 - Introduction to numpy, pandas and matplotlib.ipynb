### Example 2: Difference between numpy arrays and Python lists

```python
# Python list operations
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)  # Concatenates the lists

# numpy array operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # Element-wise addition
```

```python
import numpy as np
```

```python
x = np.arange(10)
np.random.seed(42)
y = np.random.random(10)

plt.figure(figsize=(5,2))
plt.plot(x, y, label="data")
plt.legend(loc=[1.02,0])

plt.xlabel("time")
# Single characters: 
# plt.xticks(np.arange(10), "ABCDEFGHIJ")
plt.xticks(np.arange(10), 
           ["baseline", "plus drug", "minus drug", "A", "B", "C", "D", "E", "F", "G"], 
           rotation=90)

# Ensures that the plot layout is decent
plt.tight_layout()
plt.savefig("plot.png", dpi=300)
```

### Example 2: Scatter Plot

```python
plt.figure(figsize=(2,2))
c = np.random.random(10)
# C is color, s is dot size
plt.scatter(x, y, c=c, s=4)
```

### Example 3: Bar Plot

```python
plt.figure(figsize=(2,2))
plt.bar([0,1,2], [5, 2, 3])
plt.xticks([0,1,2], ["A", "B", "C"])
```

### How to get these things into PowerPoint and Inkscape/Illustrator

```python
plt.rcParams['svg.fonttype'] = 'none' # Text is not rendered
plt.rcParams['pdf.fonttype'] = 42 # TrueType to avoid PDF issues
```

```python
plt.figure(figsize=(2,2))
plt.bar([0,1,2], [5, 2, 3])
plt.xticks([0,1,2], ["A", "B", "C"])
plt.savefig("barplot.svg")
plt.savefig("barplot.pdf")
plt.savefig("barplot.eps")
```
