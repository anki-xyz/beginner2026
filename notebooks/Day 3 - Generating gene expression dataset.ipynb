```python
import pandas as pd
import numpy as np

# Number of genes and samples
num_genes = 100
samples = ["Healthy_Control_1", "Healthy_Control_2", 
           "Disease_State_1", "Disease_State_2", 
           "Post_Treatment_1", "Post_Treatment_2"]

# Generate gene names
genes = [f"Gene{i}" for i in range(num_genes)]

# Generate synthetic expression data
np.random.seed(42)  # for reproducibility

# Healthy controls will have baseline expression levels with small variations
healthy_control_1 = np.random.normal(5, 0.5, num_genes)
healthy_control_2 = np.random.normal(5, 0.5, num_genes)

# For disease states, let's assume half the genes are upregulated and half are downregulated
disease_state_1_upregulated = np.random.normal(7, 0.5, num_genes // 2)
disease_state_1_downregulated = np.random.normal(3, 0.5, num_genes // 2)
disease_state_1 = np.concatenate([disease_state_1_upregulated, disease_state_1_downregulated])
np.random.shuffle(disease_state_1)  # shuffle to randomly distribute the up/downregulated genes

disease_state_2_upregulated = np.random.normal(7, 0.5, num_genes // 2)
disease_state_2_downregulated = np.random.normal(3, 0.5, num_genes // 2)
disease_state_2 = np.concatenate([disease_state_2_upregulated, disease_state_2_downregulated])
np.random.shuffle(disease_state_2)

# Post-treatment will have values close to healthy but not identical
post_treatment_1 = np.random.normal(5.5, 0.5, num_genes)
post_treatment_2 = np.random.normal(5.5, 0.5, num_genes)

# Create DataFrame
data = {
    "Gene": genes,
    "Healthy_Control_1": healthy_control_1,
    "Healthy_Control_2": healthy_control_2,
    "Disease_State_1": disease_state_1,
    "Disease_State_2": disease_state_2,
    "Post_Treatment_1": post_treatment_1,
    "Post_Treatment_2": post_treatment_2
}

df = pd.DataFrame(data)

# Save as CSV
file_path = "gene_expression_data.csv"
df.to_csv(file_path, sep=";", index=False)
```
