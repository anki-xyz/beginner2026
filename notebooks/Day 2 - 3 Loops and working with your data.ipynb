# Loops

In _Python_ there are basically two kinds of loops: `while` and `for` loops.

With `while` you repeat the code over and over again until the __loop condition__ becomes `False`.

In contrast, `for` loops go through elements of collections #

```python
data = [0, 1, 2, 3, 2, 1, 0, -1, -2, -3, -2, -1]
```

```python
data
```

```python
i = 0

while i < 5:
    # Printing i to the system  NOT SO GOOD COMMENT
    print(i)
    
    # Skipping each empty row  GOOD COMMENT
    i += 2
    
    # Inside the loop
    print("inside")
    
# Outside the loop
print("outside")
```

```python
len(data)
```

```python
data
```

```python
i = 0.
## Index always has to be an integer,
# if it is not an int, then you need to convert it

while i < len(data):
    # Indexing data at location i
    print(i, data[int(i)])
    i += 1
```

```python
while True:
    ## run experiment
    
    if time runs out:
        break
```

```python
while observations < target:
    # run experiment
```

We can do exactly the same using `for` loops.

```python
for el in data:
    print(el)
```

The next logical step would be to use the **genetic code** to translate codons to amino acids.

What we are basically doing is looking which **key**, i.e. the codon, corresponds to which amino acid, i.e. **value**.

The perfect Python collection here would be a **dictionary**!

```python
genetic_code = { 
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K', 
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                  
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L', 
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q', 
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V', 
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G', 
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L', 
        'TAC':'Y', 'TAT':'Y', 'TAA':'*', 'TAG':'*', 
        'TGC':'C', 'TGT':'C', 'TGA':'*', 'TGG':'W', 
    } 
```

```python
genetic_code['GAG']
```

The **list** approach

```python
seq = []

for i in range(0, len(gfp), 3):
    codon = gfp[i:i+3]
    
    aa = genetic_code[codon]
    
    seq.append(aa)
    
    print(i, codon, aa, seq)
```

```python
# Turn a list into a string
"".join(seq)
```

The **string** approach

```python
# Initialize the variable as a string, keep it empty because there's nothing there yet
seq = ""

for i in range(0, len(gfp), 3):
    codon = gfp[i:i+3]
    
    aa = genetic_code[codon]
    
    # Take the old sequence, add the current aa and save it again in seq
    seq = seq+aa
    
    print(i, codon, aa, seq)
```
