Can we calculate the GC content here?

$$ gccontent = \frac{Gs+Cs}{As + Ts + Gs + Cs} \cdot 100\%$$

````python
print(gfp)
````

````python
gfp.count("T")
````

````python
gfp.find("TAA")
````

````python
Cs = gfp.count("C")
Gs = gfp.count("G")
As = gfp.count("A")
Ts = gfp.count("T")
````

````python
gccontent = (Gs+Cs) / (Cs+Gs+As+Ts) * 100
gccontent
````

````python
L = len(gfp)
````

````python
(Gs+Cs) / L * 100
````

Check if everything is correctly a DNA sequence

````python
Cs+As+Gs+Ts == L
````
