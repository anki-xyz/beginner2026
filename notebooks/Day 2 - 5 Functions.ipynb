## Functions that returns something!

Another example would be a square root function!

Babylon Algorithm:

1. You are looking for the square root of $x$.
2. You have an initial estimate $e$ of $\frac{x}{2}$.
3. A better guess would be the mean of $e$ and $\frac{x}{e}$.
4. Iterate the process until you are satisfied with the accuracy of your guess.

Let's create a pseudo-code on the whiteboard and then implement it in Python.

--------------

The easiest way to compute it in Python:

````python
10**0.5
````

````python
#### argument = 5 --> default value
def sqrt(x, iterations=5):
    e = x / 2
    
    for i in range(iterations):
        e = (x/e + e) / 2
        print(i, e)
    
    return e # our estimate
    
````

````python
sqrt(329843)
````

````python
def sqrt(x):
    e = x / 2
    delta = 1
    
    while delta:
        old_e = e
        e = (x/e + e) / 2
        
        delta = e - old_e
        
        print(e, delta)
    
    return e

sqrt(1000)
````

````python
def sqrt(x, max_iterations=20):
    e = x / 2
    delta = 1
    i = 0
    
    while delta:
        old_e = e
        e = (x/e + e) / 2
        
        delta = e - old_e
        
        print(i, e, delta)
        
        i += 1
        
        if i == max_iterations:
            break        
    
    return e

sqrt(1000098790000)

````

````python
data = list([2,62,26,43,2,5,2,6,327,5,3,5,73,32,55,53,254,2])
data
````

````python
## Moving average across my data --> box filter [0, 1, 3, 1] --> [0, 1.33, 1.66, 1]

def filt(x):
    
    # We would like to keep the first element --> edge case
    x_filtered = [x[0]]
    
    # Cannot go "before" the data, start one later
    for i in range(1, len(x)):
#         print(x[i-1:i+2])
        
        # Computing the average
        avg = sum(x[i-1:i+2]) / 3
        
        x_filtered.append(avg)
    
    return x_filtered

filt(data)
````

````python
def gc_content(seq):
    gc = (seq.count('G')+seq.count('C'))/len(seq)*100
    return gc
    
````

````python
def slidingGC(seq, window=10):
    all_gcs = []
    
    for i in range(len(seq)-window):
        subset = seq[i:i+window]
        gc = gc_content(subset)
        all_gcs.append(gc)
        
    return all_gcs
    
````

````python
r = slidingGC(gfp)
````
