# Information

This is to teach you how to open, read and write files.

However, many libraries, such as `pandas`, allow easier access to e.g. CSV files!

## Write to a file

```python
with open("myfile.png", "w") as fp:
    for i in range(10):
        fp.write(f"{i} Hi there hello! \n")
```

```python
with open("math.png", "rb") as fp:
    print(fp.read())
```

## How to read a file

You can also read in line by line and then process the line directly.

# A CSV file

A CSV File contains **comma separated values**

e.g. for ID, Name, Age, Country

```python
with open("mycsv.csv","w") as fp:
    fp.write("0,person A, 23, USA\n")
    fp.write("0,person B, 34, Italy\n")
    fp.write("0,person C, 45, Sweden\n")
```
