# Conditional statements

Classic notation: __IF THEN ELSE__

"For instance, __if__ a sausage _is a fruit_ __then__ eat it __else__ cook it and then it eat."

"__if__ x is an odd number, __then__ print it, __else__ do nothing"

## Reserved words in Python

`if` __if__ this statement is true, execute what comes below

`elif` if the previous __if__ condition is false another check, short for __else if__

`else` if all __if__s and __elif__s are not true

```python
data = [0, 1, 2, 5, 3, 1]
```

A single block

```python
if data[4] > 1:
    print("larger!")
    
elif data[4] < 4:
    print("smaller than 4")
    
else:
    print("something else")
```

Two blocks

```python
if data[4] > 1:
    print("larger!")
    
if data[4] < 4:
    print("smaller than 4")
    
else:
    print("something else")
```

```python
for el in data:
    print(el)
    
    if el > 2:
        print("larger!")
    
    # Only else if you need another event/operation
    else: 
        print("something else")
```

```python
result = []

for i in range(20):
    # %2 == 0 --> even numbers
    # %2 or %2 == 1 --> odd numbers
    if i % 2 == 0:
        result.append(i)
        
result
```

## Another example

Imagine, we have a time series of some data, and NaN (not a number) values are set to -100. However, we would like to have them being 0 and not -100. How do we do this?

```python
data = [2.3, 10.2, -100, .2, .10, -100, 2.32, 3.12]
# target: data = [2.3, 10.2,    0, .2, .10,    0, 2.32, 3.12]
```

```python
data_clean = []

for i in range(len(data)):
    print(i) # i is my index
    
    content = data[i]
    print(content)
    
    if content == -100:
        data_clean.append(0)
        
    else:
        data_clean.append(content)
        
data_clean
        
```

# Combine if statements

```python
for i in range(10):
    print(i)
    
    # Nested if clauses/statements
    if i > 5:
        #print("====")
        
        if i < 7:
            print("****")
```

```python
for i in range(10):
    print(i)
    
    # Logical AND
    if i > 5 and i < 7:
        print("****")
```

```python
for i in range(10):
    print(i)
    
    # Logical OR
    if i > 5 or i < 7:
        print("****")
```

# List comprehension

A fast way to write a for loop in a single line.

https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions

```python
data = [0, 1, 2, 3, 4, 5]
```

Square each element in data, save it to a new data `list`

```python
[i for i in range(5)]
```

```python
new_data = [i**2 for i in data]

new_data
```

```python
# Selecting for a subset, if at the end
[i for i in data if i > 2]
```

```python
# Two different treatment strategies, if/else before "for"
[i if i < 3 else i**2 for i in data]
```

Do the same as above, **but only for** `element`s that are even!

```python
[i**2 if i%2==0 else i for i in data]
```
